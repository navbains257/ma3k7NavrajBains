```python
import random

def drawSubtract(arr):
    # Continue the process until only one element is left in the array
    while len(arr) > 1:
        # Draw two random indices
        index1, index2 = random.sample(range(len(arr)), 2)

        # Get the corresponding values
        num1, num2 = arr[index1], arr[index2]

        # Remove the drawn elements
        arr.pop(max(index1, index2))
        arr.pop(min(index1, index2))

        # Calculate the difference and add it to the array
        diff = abs(num1 - num2)
        arr.append(diff)

    return arr[0]

# Create an array containing numbers from 1 to 2024
# numbers = list(range(1, 2025))

# Call the function with the array
results = []
for i in range(100000):
    numbers = list(range(1, 2025))
    result = drawSubtract(list(range(1, 2025)))
    results.append(result)
    i+=1

# Print the final result
# print("The remaining number is:", result)
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample array of numbers


# Create a histogram
sns.histplot(results, bins=np.arange(min(results), max(results) + 1.5) - 0.5, edgecolor='black', alpha=0.7)

# Set labels and title
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.title('Frequency of Numbers in Array')


# Show the plot
plt.show()
print(results)
```

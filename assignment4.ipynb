```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from scipy import linalg as LA
from sympy.abc import m

M = sp.solve(m**2 -0.5*m - 0.5, m)
# print(M)

x,y = sp.symbols('x,y')
A = sp.Matrix([[1,1],[M[0], M[1]]])
b = sp.Matrix([0,1])
C, = sp.linsolve((A,b), x,y)
# print(C)

qVals= [0.2,0.4,0.5, 0.6, 0.8]
x = np.arange(0,25)
plt.figure(figsize=(10, 6))
for q in qVals:
    y = (1-(q-1)**x)/(2-q)
    plt.plot(x,y, 'x', linestyle='-', markersize=5, label=f'q={q}')
    plt.xlabel('p (Number on path)')
    plt.ylabel(r'$\mathbb{P}(p)$')
    plt.title('Probability of landing on step p at some point')
    plt.legend()
plt.show()
```

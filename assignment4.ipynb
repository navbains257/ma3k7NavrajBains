```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from scipy import linalg as LA
from sympy.abc import m

M = sp.solve(m**2 -0.5*m - 0.5, m)
# print(M)

x,y = sp.symbols('x,y')
A = sp.Matrix([[1,1],[M[0], M[1]]])
b = sp.Matrix([0,1])
C, = sp.linsolve((A,b), x,y)
# print(C)

x = np.arange(0,26)
y = 2/3 * (1- (-1/2)**x)
plt.plot(x,y, 'x', color='r', markersize=5)
plt.xlabel('p (Number on path)')
plt.ylabel(r'$\mathbb{P}(p)$')
plt.axhline(y=2/3, color='g', linewidth=0.5)
plt.title('Probability of landing on step p at some point')
plt.savefig('prob25.png')

x = np.arange(0,100)
y = 2/3 * (1- (-1/2)**x)
plt.plot(x,y, 'x', color='r', markersize=5)
plt.xlabel('p (Number on path)')
plt.ylabel(r'$\mathbb{P}(p)$')
plt.title('Probability of landing on step p at some point')
plt.axhline(y=2/3, color='g', linewidth=0.5)
plt.savefig('prob100.png')
```

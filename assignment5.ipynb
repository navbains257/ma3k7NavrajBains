```python
import numpy as np
import matplotlib.pyplot as plt

# Provided function to compute coordinates
def coordinate(n):
    s = int(np.ceil(np.sqrt(n)))
    if s % 2 == 0:
        s += 1
    twoL = s - 1
    if twoL == 0:
        r = ((twoL + 1) ** 2 - n)
        m = 0
    else:
        r = ((twoL + 1) ** 2 - n) % twoL
        m = ((twoL + 1) ** 2 - n - r) / twoL
    L = twoL / 2
    if m == 0:
        coords = (L - r, -L)
    elif m == 1:
        coords = (-L, -L + r)
    elif m == 2:
        coords = (-L + r, L)
    else:
        coords = (L, L - r)
    return coords

N = 49

coordinates = [coordinate(n) for n in range(1, N+1)]

x_coords, y_coords = zip(*coordinates)

x_min, x_max = min(x_coords), max(x_coords)
y_min, y_max = min(y_coords), max(y_coords)

fig_width = max(x_max - x_min, y_max - y_min) * 1.2
fig_size = max(6, fig_width) 

plt.figure(figsize=(fig_size, fig_size))

for i, (x, y) in enumerate(coordinates, start=1):
    plt.text(x, y, str(i), fontsize=9, ha='center', va='center')

plt.xlim(x_min - 1, x_max + 1)
plt.ylim(y_min - 1, y_max + 1)

plt.axis('off')
plt.tight_layout()
plt.show()
```
